```python
import numpy as np
import random
import math

class Graph:
    def __init__(self):
        self.vertices = {}
        self.adjacent = {}
        
    def add_vertex(self, name, pos):
        self.vertices[name] = pos
        self.adjacent[name] = []
        
    def add_edge(self, v1, v2):
        if v1 not in self.vertices or v2 not in self.vertices:
            raise ValueError("Both vertices must be in the graph.")
        if v2 not in self.adjacent[v1]:
            self.adjacent[v1].append(v2)
        if v1 not in self.adjacent[v2]:
            self.adjacent[v2].append(v1)
            
    def remove_edge(self, v1, v2):
        self.adjacent[v1].remove(v2)
        self.adjacent[v2].remove(v1)
        
    def get_neighbors(self, v):
        return self.adjacent[v]
    
    def get_vertices(self):
        return self.vertices.keys()
    
    def dist(self, v1, v2):
        return np.linalg.norm(self.vertices[v1] - self.vertices[v2])
    
def naive_knn(g, v, k):
    result = []
    for v2 in g.get_vertices():
        if v2 != v:
            result.append(v2)
            result.sort(key=lambda x: g.dist(v, x))
            if len(result) > k:
                result = result[:k]
    return result

def insert_vertex(g, v, pos, k):
    g.add_vertex(v, pos)
    nn = naive_knn(g, v, k)
    for v2 in nn:
        g.add_edge(v, v2)
    for v2 in list(g.get_neighbors(v)):
        neighbors = list(g.get_neighbors(v2))  # Make a copy of the list
        if len(neighbors) > k:
            # Sort the neighbors by distance to v2
            sorted_neighbors = sorted(neighbors, key=lambda x: g.dist(x, v2))
            # Remove the edges to neighbors beyond the k-th neighbor
            for v3 in sorted_neighbors[k:]:
                g.remove_edge(v2, v3)

class HNSW:
    def __init__(self):
        self.layers = []
        
    def add_vertex(self, name, pos):
        m_L = 0.6
        layer = math.floor(-math.log(random.random())*m_L)
        degree_higher_layers = 4
        degree_layer_0 = 8
        while len(self.layers) <= layer:
            self.layers.append(Graph())
        for i in range(layer+1):
            if i == 0:
                d = degree_layer_0
            else:
                d = degree_higher_layers
            insert_vertex(self.layers[i], name, pos, d)
        
    


        
        
```

```python
import matplotlib.pyplot as plt

def plot_graph(graph):
    """
    Plots the graph using matplotlib.

    Parameters:
    graph (Graph): An instance of the Graph class.
    """
    
    plt.figure(figsize=(12, 12))
    # Plot the vertices
    for v, pos in graph.vertices.items():
        plt.scatter(pos[0], pos[1], color='blue', zorder=2)
        #plt.text(pos[0], pos[1], f'{v}', fontsize=12, ha='right', va='bottom')

    # Plot the edges
    plotted_edges = set()
    for v1, neighbors in graph.adjacent.items():
        for v2 in neighbors:
            # Ensure each edge is plotted only once
            edge = tuple(sorted((v1, v2)))
            if edge not in plotted_edges:
                pos1 = graph.vertices[v1]
                pos2 = graph.vertices[v2]
                plt.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]], color='black', zorder=1)
                plotted_edges.add(edge)

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.xlim((0, 10))
    plt.ylim((0, 10))
    plt.title('Graph Plot')
    plt.grid(False)
    plt.axis('equal')
    plt.show()
    
def plot_hnsw(h):
    """
    Plots the graph using matplotlib.

    Parameters:
    graph (Graph): An instance of the Graph class.
    """
    
    colors = ["gray", "green", "red", "orange", "pink"]
    
    plt.figure(figsize=(12, 12))
    # Plot the vertices
    for v, pos in h.layers[0].vertices.items():
        plt.scatter(pos[0], pos[1], color='blue', zorder=2)
        #plt.text(pos[0], pos[1], f'{v}', fontsize=12, ha='right', va='bottom')

    # Plot the edges
    for i, graph in enumerate(h.layers):    
        plotted_edges = set()
        for v1, neighbors in graph.adjacent.items():
            for v2 in neighbors:
                # Ensure each edge is plotted only once
                edge = tuple(sorted((v1, v2)))
                if edge not in plotted_edges:
                    pos1 = graph.vertices[v1]
                    pos2 = graph.vertices[v2]
                    plt.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]], color=colors[i], zorder=1)
                    plotted_edges.add(edge)

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.xlim((0, 10))
    plt.ylim((0, 10))
    plt.title('Graph Plot')
    plt.grid(False)
    plt.axis('equal')
    plt.show()    
```

```python
h = HNSW()
for i in range(300):
    pos = np.random.rand(2)*10
    h.add_vertex(i, pos)
plot_hnsw(h)
```
